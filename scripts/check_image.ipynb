```python
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_npy_image(npy_file: Path):
    """
    npyファイルをプロットする

    Args:
        npy_file (Path): npyファイルのパス
    """
    image = np.load(npy_file)
    plt.imshow(image)
    plt.show()
```

```python
plot_npy_image(
    "../data/atmaCup#18_dataset/images/00bb6daff4f6a323918044e01651410d_420/depth_image_t.png.npy"
)
```
